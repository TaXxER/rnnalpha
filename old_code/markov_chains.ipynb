```python
import numpy as np
from sklearn.metrics import brier_score_loss
```

```python
traces = []
with open("Repos/rnnalpha/textual_logs/sepsis.txt", "r") as fin:
    for line in fin:
        traces.append(["START"] + list(line.strip()) + ["END"])
```

```python
traces
```

```python
vocabulary = set([val for trace in traces for val in trace])
vocabulary = {key: idx for idx, key in enumerate(vocabulary)}
```

```python
vocabulary
```

```python
markov_chain = np.zeros((len(vocabulary), len(vocabulary)))
```

```python
for trace in traces:
    for i in range(len(trace)-1):
        markov_chain[vocabulary[trace[i]], vocabulary[trace[i+1]]] += 1
```

```python
markov_chain = (markov_chain.T / markov_chain.sum(axis=1)).T
```

```python
markov_chain = np.delete(markov_chain, vocabulary["START"], axis=1) 
```

```python
scores = []
for trace in traces:
    for i in range(len(trace)-1):
        predicted = markov_chain[vocabulary[trace[i]]]
        actual = np.zeros(len(vocabulary))
        actual[vocabulary[trace[i+1]]] = 1
        actual = np.delete(actual, vocabulary["START"]) 
        score = brier_score_loss(actual, predicted)
        scores.append(score)
```

```python
np.mean(scores)
```
